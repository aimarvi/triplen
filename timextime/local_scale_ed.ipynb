```python
import os, sys, pickle

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from tqdm import tqdm
import manifold_dynamics.tuning_utils as tut
```

```python
DATA_DIR = '../../datasets/NNN/'
# IMG_DIR = '../../datasets/NNN/NSD1000_LOC'

CAT = 'face'
dat = pd.read_pickle(os.path.join(DATA_DIR, (f'{CAT}_roi_data.pkl')))
ROI_LIST = list(dat['roi'].unique())
print(f'Unique {CAT} ROIs: {ROI_LIST}')

# optimal k: the optimal manifold scale at which we should conduct analyses
# see ./pause.ipynb
with open(f'../../datasets/NNN/{CAT}_mins.pkl', 'rb') as f:
    mins = pickle.load(f)
SCALES = {k: v[0] for k,v in mins.items()}
```

```python
DATA_DIR = '../../datasets/NNN/'
# IMG_DIR = '../../datasets/NNN/NSD1000_LOC'

CAT = 'object'
obj_dat = pd.read_pickle(os.path.join(DATA_DIR, (f'{CAT}_roi_data.pkl')))
ROI_LIST = list(obj_dat['roi'].unique())
print(f'Unique {CAT} ROIs: {list(obj_dat['roi'].unique())}')

scores = tut.landscape(obj_dat, rng.choice(list(obj_dat['roi'].unique())))
obj_order = np.argsort(scores)[::-1]
```

```python
ROI = 'MF1_8_F'  # Unknown_19_F, MF1_7_F, MF1_8_F, MF1_9_F, AF3_18_F, MB1_3_B

tstart = 100
tend=350
scale = SCALES[ROI]

# find top images
rng = np.random.default_rng(0)
scores = tut.landscape(dat, ROI)
order = np.argsort(scores)[::-1]
# order[:scale] # rng.choice(order[scale:], scale) #order[:scale]

fig,axes = plt.subplots(1,3)
for i, img_set in enumerate([0, 24, 48]):
    ax = axes[i]
    idx = slice(1000+img_set,1000+img_set+24) 
    R, _ = tut.specific_static_rdm(dat, ROI, idx, tstart=tstart, tend=tend)
    ed = tut.ED2(R)
    sns.heatmap(R, square=True, vmin=0, vmax=1, ax=ax, cbar=False)
    ax.set_title(f'ED: {ed:.02f}')
plt.show()

fig, axes = plt.subplots(1,2)
for i, mode in enumerate(['top', 'rand']):
    ax = axes[i]
    if mode=='top': 
        idx = order[:scale]
    else:
        idx = rng.choice(order[scale:], scale)
    R, _ = tut.specific_static_rdm(dat, ROI, idx, tstart=tstart, tend=tend)
    ed = tut.ED2(R)
    sns.heatmap(R, square=True, vmin=0, vmax=1, ax=ax, cbar=False)
    ax.set_title(f'{ROI} {mode} ED: {ed:.02f}')
plt.show()    
```

```python
ROI = 'MF1_7_F'  # Unknown_19_F, MF1_7_F, MF1_8_F, MF1_9_F, AF3_18_F, MB1_3_B

tstart = 100
tend=350
scale = SCALES[ROI]

fig, axes = plt.subplots(1,3)
for i, mode in enumerate(['top', 'rand', 'alternative']):
    ax = axes[i]
    scores = tut.landscape(dat, ROI)
    order = np.argsort(scores)[::-1]
    
    if mode=='top': 
        idx = order[:scale]
    elif mode =='rand':
        idx = rng.choice(order[scale:], scale)
    else:
        idx = obj_order[:scale]
        
    R, _ = tut.specific_static_rdm(dat, ROI, idx, tstart=tstart, tend=tend)
    ed = tut.ED2(R)
    sns.heatmap(R, square=True, vmin=0, vmax=1, ax=ax, cbar=False)
    ax.set_title(f'{ROI} {mode}\nED: {ed:.02f}')
plt.tight_layout()
plt.show()    
```
